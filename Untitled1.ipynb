# Agrupar y agregar

```python
import pandas as pd
from google.colab import files

# Subir archivos
# Asegúrate de subir tanto 'ventas_mensuales.xlsx' como 'empleados.xlsx'
uploaded = files.upload()

# Cargar datasets
ventas = pd.read_excel('ventas_mensuales.xlsx')
empleados = pd.read_excel('empleados.xlsx')
```

```python
ventas['Total_Venta'] = ventas['Cantidad'] * ventas['Precio_Unitario']
```

```python
ventas_region = ventas.groupby('Region')['Total_Venta'].sum()

print("Total de ventas por región:")
print(ventas_region)
```

```python
ventas_region.sort_values(ascending=False)
```

```python
promedio_producto = ventas.groupby('Producto')['Cantidad'].mean()

print("Cantidad promedio por producto:")
print(promedio_producto)
```

```python
salario_promedio = empleados.groupby('Departamento')['Salario'].mean()

print("Salario promedio por departamento:")
print(salario_promedio)
```

```python
salarios_ciudad = empleados.groupby('Ciudad')['Salario'].agg(['max', 'min'])

print("Salario máximo y mínimo por ciudad:")
print(salarios_ciudad)
```

# Unión de datos

```python
import pandas as pd
from google.colab import files

# Subir archivos
uploaded = files.upload()

# Cargar datasets
ventas = pd.read_excel('ventas_mensuales.xlsx')
productos = pd.read_excel('productos.xlsx')

# Ver datos
print(ventas.head())
print(productos.head())
```

```python
# Crear SKU simplificado en ventas
ventas['SKU'] = ventas['Producto'].str.upper().str.replace(" ", "_")
```

```python
productos['SKU'] = productos['SKU'].str.upper().str.replace(" ", "_")
```

```python
df_combinado = pd.merge(ventas, productos, on='SKU', how='left')
```

```python
df_combinado = pd.merge(ventas, productos, on='SKU', how='left')
```

```python
ventas_por_producto = df_combinado.groupby('SKU')['Cantidad'].sum().reset_index()
ventas_por_producto.rename(columns={'Cantidad': 'Total_Vendido'}, inplace=True)
```

```python
df_stock = pd.merge(productos, ventas_por_producto, on='SKU', how='left')
```

```python
df_stock['Total_Vendido'] = df_stock['Total_Vendido'].fillna(0)
```

```python
df_stock['Total_Vendido'] = df_stock['Total_Vendido'].fillna(0)
```

```python
df_stock['Stock_Restante'] = df_stock['Stock_Inicial'] - df_stock['Total_Vendido']
```

# Pipeline

```python
import pandas as pd
from google.colab import files

# =========================
# 0) Upload (si estás en Colab)
# =========================
uploaded = files.upload()

# =========================
# 1) EXTRACT
# =========================
ventas = pd.read_excel("ventas_mensuales.xlsx")

# =========================
# 2) TRANSFORM
# =========================

# 2.1 Calcular Total_Venta
ventas["Total_Venta"] = ventas["Cantidad"] * ventas["Precio_Unitario"]

# 2.2 Filtrar solo ventas mayores a 1000
ventas_filtradas = ventas[ventas["Total_Venta"] > 1000].copy()

# 2.3 Agrupar por Región y sumar ventas
reporte_regional = (
    ventas_filtradas
    .groupby("Region", as_index=False)["Total_Venta"]
    .sum()
    .rename(columns={"Total_Venta": "Ventas_Totales"})
)

# 2.4 Ordenar de mayor a menor
reporte_regional = reporte_regional.sort_values("Ventas_Totales", ascending=False)

# =========================
# 3) LOAD
# =========================
reporte_regional.to_excel("reporte_regional.xlsx", index=False)

# (Opcional) Ver resultado en pantalla
print("Reporte regional (ventas > 1000):")
print(reporte_regional)

# (Opcional) Descargar archivo generado
files.download("reporte_regional.xlsx")
```
